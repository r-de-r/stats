2 変量のグラフ表示では散布図が主なものである。

## 例示に使用するデータセット

### カテゴリーデータ

HairEyeColor データセットは，学生の髪と目の色と性別の集計表である。以下に示すのは男子学生の集計結果を示す最初の 16 行である。

```julia
using RDatasets
using FreqTables
haireyecolor = dataset("datasets", "HairEyeColor")
haireyecolor[!, 1:3] = string.(haireyecolor[!, 1:3]) # String7 が使えないパッケージに対する過渡的な対処法
size(haireyecolor) |> println
first(haireyecolor, 16)
```

例えば，男子学生で目の色が Brown の場合に，Hair の色の度数を抽出すれば 1 つの棒グラフを描くことができるが，目の色が Blue の場合は？さらに 女子学生の場合は？と，描く棒グラフが増えてくると煩雑になる。

そこで，この集計表に基づいて，元のデータセットつまり学生一人ずつのデータセットを作成しておくとよい。

```julia
HEC = DataFrame()
for row in eachrow(haireyecolor), j in 1:row.Freq
    push!(HEC, row)
end
select!(HEC, Not(:Freq)); # Freq 列を除く
```

```julia
first(HEC, 2)
```

```julia
last(HEC, 2)
```

## 棒グラフ

`bar()` は，度数分布をグラフにするので，データフレーム上の変数の棒グラフを描くにはまず度数分布表をつくらなければならない。

データ全体を対象にする場合には 1 次元の度数分布表，グループを考慮する場合には 2 次元の度数分布表を作る。

### 一標本の場合

男女込みで Hair の度数分布を求める。

```julia
freq = freqtable(HEC.Hair)
```

オプションを何も指定しないと不適切な棒グラフしか得られないことが多い。

```julia
bar(freq)
```

```julia
bar(freq,
    bar_width=1,
    xlabel="Hair color",
    xticks=(1:4, names(freq)[1]), # カテゴリーの名前
    ylabel="Number of students",
    title="Distribution of hair color in students.",
    )
```

`direction=:horizontal` を指定すると水平棒グラフになる。$x$ 軸と $y$ 軸に与える内容が入れ替わるので注意が必要である。カテゴリー名が長い場合には水平棒グラフのほうが適していることもある。

```julia
bar(freq,
    direction=:horizontal,
    bar_width=1,
    ylabel="Hair color",
    yticks=(1:4, names(freq)[1]), # カテゴリーの名前
    xlabel="Number of students",
    title="Distribution of hair color in students.",
    )
```

### 二標本以上の場合

二次元集計表により，列ごとの棒グラフを描くことができる。

```julia
freq2 = freqtable(HEC.Hair, HEC.Sex)
```

#### 横に並べる棒グラフ

複数の標本での棒グラフを 1 つにまとめて表示する方法としては 2 通りある。

1 つは，それぞれのカテゴリーでの棒を横に並べる方法である。

```julia
using StatsPlots
p1 = groupedbar(freq2,
    bar_width=0.8,
    bar_position = :dodge,
    xlabel = "Hair　color",
    xticks = (1:4, names(freq2)[1]),
    ylabel = "Number of students",
    label = ["Female" "Male"],
    title = "Distribution of hair color in students.",
    )
```

```julia
p2 = groupedbar(freq2,
    bar_width=0.8,
    bar_position = :dodge,
    direction=:horizontal,
    ylabel = "Hair　color",
    yticks = (1:4, names(freq2)[1]),
    xlabel = "Number of students",
    label = ["Female" "Male"],
    title = "Distribution of hair color in students.",
    )
```

#### 積み上げ棒グラフ

```julia
using StatsPlots
p3 = groupedbar(freq2,
    bar_width=0.8,
    bar_position = :stack,
    xlabel = "Hair color",
    xticks = (1:4, names(freq2)[1]),
    ylabel = "Number of students",
    label = ["Female" "Male"],
    title = "Distribution of hair color in students.",
    )
```

```julia
using StatsPlots
p4 = groupedbar(freq2,
    bar_width=0.8,
    bar_position = :stack,
    direction=:horizontal,
    ylabel = "Hair color",
    yticks = (1:4, names(freq2)[1]),
    xlabel = "Number of students",
    label = ["Female" "Male"],
    title = "Distribution of hair color in students.",
    )
```

### 複数のグラフを行列状にまとめて表示する方法

前節 1.2.1., 1.2.2. で描いた 2 つのグラフを1つにまとめることができる。

`plot()` の引数として，描かれたグラフを指定する。

描かれるグラフはオブジェクトという変数のようなものに代入できる。実際，`p1 = groupedbar(freq,...)`，`p2 = groupedbar(freq,...)` のように，`p1`, `p2` に代入した。

このオブジェクトを `plot()` の引数とすればよい。

```julia
using Plots.PlotMeasures # 余白の単位 mm などを有効にするために必要
plot(p1, p2, p3, p4,
    legend=:topright, # 凡例の位置 :topright, :bottomleft などなど
    size=(750, 550),  # 横幅，高さをピクセル単位で指定
    bottom_margin = 20px #   サブプロットのボトムの余白 xxpx
)
```

この手法を用いれば，系統的に作成したグラフをまとめて提示することができる。

```julia
freq2[:,"Male"]
```

```julia
function haircolor(sex)
    bar(
        freq2[:,sex],
        bar_width = 1,
        xlabel = "Hair color",
        xticks = (1:4, names(freq)[1]), # カテゴリーの名前
        ylabel = "Number of students",
        title = "Distribution of hair color in $(lowercase(sex)) students.",
    )
end
```

```julia
female = haircolor("Female")
```

```julia
male = haircolor("Male")
```

```julia
plot(male, female, size=(700, 350))
```

たとえば，性別に髪の色と目の色の度数分布表を作ってみよう。度数分布表は三次元になる。

```julia
freq3 = freqtable(HEC.Hair, HEC.Eye, HEC.Sex)
```

```julia
plots = Array{Any, 1}(undef, 4)
for (j, eye) in enumerate([:Blue, :Brown, :Green, :Hazel])
    plots[j] = groupedbar(
        freq3[:, j, :],
        bar_position = :dodge,
        xlabel = "Hair color",
        xticks = (1:4, names(freq3)[1]),
        titlefont = (10, "times"),
        title = "Eye:$eye",
        )
end
```

```julia
using Plots.PlotMeasures # 余白の単位 mm などを有効にするために必要
plot(plots[1], plots[2], plots[3], plots[4], 
     layout=(2,2),
     bottom_margin = 20px #   サブプロットのボトムの余白 xxpx
)
```

## 帯グラフ

帯グラフは，グループごとの相対度数分布を積み上げ棒グラフにしたものなので，以下のような関数を作っておけばよい。

[プログラムダウンロード](obi.jl)

```julia
include("../../Documents/GitHub/stats/obi.jl");
```

使用法は，`freqtable()` の第 1 引数にグループ分けの変数，第 2 引数に分布を表示する変数として得た二重クロス表を `barplot2()` に与える。

```julia
freq2 = freqtable(HEC.Hair, HEC.Sex)
obi(freq2, ylabel="Hair", title="Distribution of hair color in students")
```

```julia
obi(freqtable(HEC.Sex, HEC.Hair), title="Distribution of hair color in students")
```
